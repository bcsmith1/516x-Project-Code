```python
# Input the Fan RPM data file, clean, average and merge the two data files into 1 file
drpm1= pd.read_csv("10-10-19RPM1.csv", header=1)
drpm1.head()
```

```python
drpm2= pd.read_csv("10-10-19RPM2.csv", header=1)
drpm2.head()
```

```python
drpm1.columns = ['number', 'Date_Time','RPM']
drpm13 = drpm1.drop(['number'], axis=1)
drpm2.columns = ['number', 'Date_Time','RPM']
drpm23 = drpm2.drop(['number'], axis=1)
rpm1 = pd.merge(left=drpm13, right=drpm23, how='left', left_on='Date_Time', right_on='Date_Time')
rpm1.head()
```

```python
#replace 0's with NAN then replace NAN with the averages
rpm2=rpm1.replace(0, np.nan)
rpm2.head()
```

```python
#fill missing values with the average
rpm3 = rpm2.copy()
rpm3['RPM_x'] = rpm2['RPM_x'].fillna(rpm2['RPM_x'].mean())
rpm3['RPM_y'] = rpm2['RPM_y'].fillna(rpm2['RPM_x'].mean())
rpm3.head()
```

```python
rpm3['Averagerpm']= (rpm3['RPM_x']+ rpm3['RPM_y'])/2
rpm3.head()
```

```python
rpm3.to_csv('RPM_processed.csv')
```

```python
#read in the manually fixed file for the missing values
df21 = pd.read_excel('BNW.xlsx', sheet_name='BNW')
df21.head()
```

```python
#renamed the columns
df21.columns = ['station', 'Date_Time','Tamb','RHamb','winddir','windspd']
df121a = df21.drop(['station'], axis=1)

```

```python
# fill missing value interpolation method
df121a['Tamb'].interpolate(method='linear', axis=0)
df121a['RHamb'].interpolate(method='linear', axis=0)
df121a['winddir'].interpolate(method='linear', axis=0)
df121a['windspd'].interpolate(method='linear', axis=0)
df121a.head()
```

```python
# switch to DateTime index to up sample the data from 20 minute to 1 mintue samples with interpolate methods
df121b=df121a.set_index(pd.DatetimeIndex(df121a.Date_Time))
df121d=df121b.resample('1T').interpolate(method='linear', axis =0)
#df121c['Tamb'].interpolate(method='linear', axis=0)
#df121c['RHamb'].interpolate(method='linear', axis=0)
#df121c['winddir'].interpolate(method='linear', axis=0)
#df121c['windspd'].interpolate(method='linear', axis=0)
df121e = df121d.drop(['Date_Time'], axis=1)
df121e.head()
```
